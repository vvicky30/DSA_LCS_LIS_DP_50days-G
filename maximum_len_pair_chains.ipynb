# maximum_length of pair-chain problem statement:
### Coding Exercise: Max Length of Pair Chain
### You are given an array of n pairs pairs where pairs[i] = [lefti, righti] and lefti < righti.

### A pair p2 = [c, d] follows a pair p1 = [a, b] if b < c. A chain of pairs can be formed in this fashion.

### Return the length longest chain which can be formed.

- You do not need to use up all the given intervals. You can select pairs in any order.



### Example 1:

- Input: pairs = [[2,3],[3,4],[4,5]]
- Output: 2
- Explanation: The longest chain is [2,3] -> [4,5].

# maximum_length of pair-chain problem question [1-d table Tabulation(bottom-up)approach] [code-implementation]:



```python
def findLongestChain(pairs):
    n = len(pairs)  # length of array of pairs 
    pairs.sort()  # first sort the array of pairs [either based on their 1st element or last element respectively] 
    dp = [1]*n   # intialised 1-d dp_table with '1' value
    max = 1  # as we intialised above 1-d dp_table with '1' so here we also initialised max =1 for comparison purpose
    for i in range(1,n): # i from 1 to n 
        for j in range(i): # j will be goes from zero upto the i only   [for each instance or iteration of i] 
            if pairs[j][1] < pairs[i][0] and dp[j]+1>dp[i]:    # formula (a2 < b1)  and dp[j]+1 > dp[i] 
                dp[i]=dp[j]+1  # then include the pair and update dp[i] to the dp[j]+1
        if dp[i]> max:  # after filling succesivly fiiling and updating dp[i] ; iteratively with i we will update max value to value at dp[i]
                       #(if current value at dp[i] greater than last saved max value) 
            max = dp[i]
    return max      # return maximum value from dp_table as that will be the longest or maximum_length of pair-chain

# example usage :
pairs1 = [[1,2],[2,9],[4,5]]
print(findLongestChain(pairs1))  # should be 2  i.e. [1,2][4,5]

pairs2 = [[2,3],[3,4],[4,5]]   
print(findLongestChain(pairs2))# should be 2  i.e. [2,3][4,5]
```

```python
def findLongestChain(pairs):
    n = len(pairs)
    pairs.sort()
    dp = [1]*n
    res = 1
    for i in range(1,n):
        for j in range(i):
            if pairs[j][1] < pairs[i][0] and dp[j]+1>dp[i]:
                dp[i]=dp[j]+1
        if dp[i]>res: res = dp[i]
    return res            
```
