# Coding Exercise: LCS ( Longest Common Subsequence)
### Given two strings text1 and text2, return the length of their longest common subsequence. If there is no common subsequence, return 0.

### A subsequence of a string is a new string generated from the original string with some characters (can be none) deleted without changing the relative order of the remaining characters.

- For example, "ace" is a subsequence of "abcde".

- A common subsequence of two strings is a subsequence that is common to both strings.

- Example: 

Input: text1 = "pbcdq", text2 = "pcq" 

Output: 3  

Explanation: The longest common subsequence is "pcq" and its length is 3.

# longest common subsequence memoisation(top-down) approach:
### for this approach we have to construct 2-d dp_table for storing the solutions of overlapping subproblems.
### here 2-d dp_table will ne of n*m dimension where n denotes length of string-1 and m denotes length string-2 
### here the yellow-arrow specified cell in the picture denoting the sub-problem where string-1 is 'ac' and string-2 is 'abc' specifically here in this cell we're matching char 'c' of string-1 with char 'c' of string-2  //which will be matched eventually in this case.  

### we will recursively compute the solutions of subproblem and filled up the cells of dp_table here,
### here solution of subproblem already calculated then dp_table's cell at that particular index is not '-1' and we will reuse anf return the value(solution) from that particular cell at subproblen indexes. 
### in this way we're not calculating solution recursively again and agian for overlapping subproblems.


# longest common subsequence memoisation(top-down) approach [complexity]:
### remmeber that here we're using 2-d Dp_table and filling up cells of dp_table using recursion call-stack 
### so Sc = space taken by Dp_Table + space taken by recursive call_stack = O(n*m) + O(n+m)  = O(n * m)   //as 'n * m' is bigger
### Here tc is optimized as we're only calulating solution of overlapping subproblems only once by storing solution to dp_table
#### so here tc = O(n*m) 



# longest common subsequence memoisation(top-down) approach [code implementation]:


```python
# code-implementation:

def longestCommonSubsequence(text1, text2):
    # as here we're gien two strings with var text-1 and tex-2
    n = len(text1)  # length of string1
    m = len(text2)  # length of string2
    #now construct 2-d dp_table witn m-columns and n-rows all cell in the table initialised with -1
    dp = [[-1]*m for _ in range(n)]
    def LCS_recursive(index1, index2): 
        #here this recursive function will take indexes of both the given string,
                #index1 for text1 and index2 for text2
        #base-cases if indexes of any of the strings goes beyond the respective last chars of strings
        if index1>n-1 or index2>m-1:
            return 0  # as no further chars to match with relatively greater length string's chars ,so no char included in subsequence.
        #checking whether out dp_table has solution to the subproblem at these indexes or not if yes then return value(solution)
        if dp[index1][index2] !=-1:
            return dp[index1][index2]  
        #recursive scenerio
        #scenerio-1  // where at both the same indexes of string chars matched, then increase both the indexes to the next
        if text1[index1] == text2[index2]:
            dp[index1][index2] = 1 + LCS_recursive(index1+1, index2+1)   # storing solution to subproblem at these indexes
            return dp[index1][index2]  # returning solution
        #else
        #scenerio-2 // where at both the same indexes of strings chars not matched, 
        #then incrase one of the index of string to the next while remained another index of string constant(same)
        # do the vice-versa of the above 
        # and choose maximum among above two-cases  //as we want longest common subsequence
        dp[index1][index2] = max(LCS_recursive(index1+1, index2), LCS_recursive(index1,index2+1))
        return dp[index1][index2]  # returning final solution
    return LCS_recursive(0,0)  # here recursive fn initialized with both indexes equals to 0
    
#example usage :
txt1 =  "pbcdqaflaz"                
txt2 =  "pcqaazex"
print(longestCommonSubsequence(txt1,txt2))  # should be 6 //pcqaaz              

tx1 =  "aloha"                
tx2 =  "ohha"    
print(longestCommonSubsequence(tx1,tx2))  # should be 3 //oha    
```

```python
def longestCommonSubsequence(text1, text2):
    n = len(text1)
    m = len(text2)
    dp = [[-1]*m for _ in range(n)]
    def helper(index1,index2):
        #base case
        if index1>n-1 or index2>m-1:
            return 0

        if dp[index1][index2]!=-1:
            return dp[index1][index2]    

        #recursive case
        if text1[index1]==text2[index2]:
            dp[index1][index2]=1+ helper(index1+1,index2+1)
            return dp[index1][index2]
        dp[index1][index2]=max(helper(index1+1,index2), helper(index1,index2+1)) 
        return dp[index1][index2]   
    return helper(0,0)    
 
#example usage :
txt1 =  "pbcdqaflaz"                
txt2 =  "pcqaazex"
print(longestCommonSubsequence(txt1,txt2))  # should be 6 //pcqaaz              
           
```
